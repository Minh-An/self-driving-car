```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
```

```python
# color thresholding in grayscale

image = mpimg.imread('test6.jpg')
thresh = (180,255)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
binary = np.zeros_like(gray)
binary[(gray > thresh[0]) & (gray <= thresh[1])] = 1

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(24,9))
fig.tight_layout()
ax1.imshow(gray, cmap='gray')
ax2.imshow(binary, cmap='gray')
```

```python
# thresholding individual RGB color channels 

for i in range(3):
    color = 'Red' if i == 0 else 'Green' if i == 1 else 'Blue'
    
    color_channel = image[:,:,i]
    color_thresh = (200,255)
    binary_color = np.zeros_like(color_channel)
    binary_color[(color_channel > color_thresh[0]) & (color_channel <= color_thresh[1])] = 1
    fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,9))
    fig.tight_layout()
    ax1.set_title(color + ' Channel')
    ax1.imshow(color_channel, cmap='gray')
    ax2.set_title(color + ' Threshold')
    ax2.imshow(binary_color, cmap='gray')
```

```python
# thresholding HLS channels 
hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)

for i in range(3):
    color = 'Hue' if i == 0 else 'Lightness' if i == 1 else 'Saturation'
    
    channel = hls[:,:,i]
    thresh = (90,255) if i == 2 else (15,100) if i == 0 else (150,255)
    binary = np.zeros_like(channel)
    binary[(channel > thresh[0]) & (channel <= thresh[1])] = 1
    fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,9))
    fig.tight_layout()
    ax1.set_title(color + ' Channel')
    ax1.imshow(channel, cmap='gray')
    ax2.set_title(color + ' Threshold')
    ax2.imshow(binary, cmap='gray')
```
